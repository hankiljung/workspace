```python
import pandas
df =pandas.read_excel('test.xlsx')
df
```

```python
# 영어점수 다 5 증가
df['영어점수'] = df['영어점수'] + 5 
```

```python
# 평균데이터 영어+수학
df['평균']=(df['국어점수']+df['영어점수'])/2
```

```python
# 컬럼추가 결과 = fail
df['결과'] = 'Fail'
df
```

```python
# 특정조건에 맞는 데이터 수정 
# 국어 영어 총합 170 이상 pass 
df.loc[df['국어점수']+df['영어점수'] >=170,'결과']='Pass'
df
```

```python
# 컬럼 삭제하기
df.drop(columns='취미')
```

```python
#학번에서 '번'지우기
df['학번'] =df['학번'].str.replace('번', '')
```
